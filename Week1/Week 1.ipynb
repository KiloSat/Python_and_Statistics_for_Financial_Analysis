## Generate new variables in Dataframe

#### We download a new dataset now containing only facebook stock data from [here](https://www.kaggle.com/aayushmishra1512/facebook-complete-stock-data2012-2020latest)

```python
fb = pd.read_csv(r'..\Week1\fb.csv')
```

```python
fb['Price1'] = fb['Close'].shift(-1) #using .shift(int) to shift the column upwards or downwards by the given int value
fb.head()
```

```python
fb['PriceDiff'] = fb['Price1']-fb['Close']
fb.head()
```

```python
fb['Return'] = fb['PriceDiff']/fb['Close']
fb.head()
```

#### List Comprehension

```python
fb['Direction'] = [1 if fb.loc[ind , 'PriceDiff']>0 else -1 for ind in fb.index ]
fb.head()
```

```python
fb['Average'] = (fb['Close']+fb['Close'].shift(1)+fb['Close'].shift(2))/3 #This is a moving average i.e we took the
                                                                          #closing price from the last 3 days and averaged it
fb.head()
```

```python
fb['MA300'] = fb['Close'].rolling(300).mean() # we use the .rolling() method for averaging out the data
fb['MA40'] = fb['Close'].rolling(40).mean()
```

```python
fb['Close'].plot() # plotting the prices 
fb['MA300'].plot() # this is a slow signal 
fb['MA40'].plot()  # this is a fast signal
```

## Trading Strategy

```python
fb['MA10'] = fb['Close'].rolling(10).mean()
```

```python
fb['Close'].plot(legend=True) # plotting the prices 
fb['MA10'].plot(legend=True) # this is a slow signal 
fb['MA40'].plot(legend=True)  # this is a fast signal
```

#### If the MA10>MA40, we buy stock and hold as we predict the stock will go up in value. This is called LONG

```python
fb['Shares'] = [1 if fb.loc[i,'MA10']>fb.loc[i,'MA40'] else 0 for i in fb.index]
fb.head()
```

```python
fb['Profit'] = [fb.loc[ei ,'Price1']-fb.loc[ei ,'Close']              # Calculating Profit 
               if fb.loc[ei, 'Shares']==1 else 0 for ei in fb.index]
fb.tail()
```

```python
fb['Profit'].plot() # Profit Distribution
```

#### Cumulative Wealth 

```python
fb['wealth'] = fb['Profit'].cumsum() # calculating total profit
fb.tail()
```

```python
fb.loc[fb.index[-2],'wealth']  # final wealth
fb.loc[fb.index[0],'Close']    # total money spent
fb['wealth'].plot()
```
